```python
from docling.document_converter import DocumentConverter
from pprint import pprint
from langchain_ollama import OllamaLLM
import os
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
model = OllamaLLM(model="llama3.2:3b")

response = ''

# Specify the directory path
directory_path = "test-files/"

# List all files in the directory
file_names = os.listdir(directory_path)

converter = DocumentConverter()

# Print each file name
for file_name in file_names:
  source = directory_path + file_name # document per local path or URL
  result = converter.convert(source)

  response += "===\n" + model.invoke("Give me concise yet a nuanced summary of this: \n" + result.document.export_to_markdown())

  pprint(response)
  # output: "## Docling Technical Report[...]"
pprint(response)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def get_download_href(url):
    """
    Fetches the Selenium WebDriver, navigates to the target URL, performs clicks and searches,
    and extracts the href attribute from the download button.

    Args:
        url (str): The URL of the webpage to interact with.

    Returns:
        str: The href attribute of the download button if found, None otherwise.
    """

    driver = webdriver.Chrome()  # Replace with your preferred WebDriver
    driver.get(url)

    # Click "Select" button
    select_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "button.drop-toggle.flat"))
    )
    select_button.click()

    # Enter "SAMPATH BANK" and submit search
    search_box = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "input[placeholder='Company Search']"))
    )
    search_box.send_keys("SAMPATH BANK")
    search_box.send_keys(Keys.ENTER)

    # Click "SAMPATH BANK PLC" checkbox
    sampath_checkbox = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//label[text()='SAMPATH BANK PLC']/input"))
    )
    sampath_checkbox.click()

    # Click "Get Data" button
    get_data_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "input[name='getData'][type='submit']"))
    )
    get_data_button.click()

    # Wait for announcement element with SAMPATH BANK PLC text
    sampath_announcement = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located(
            (By.XPATH, "//div[contains(text(), 'SAMPATH BANK PLC')]")
        )
    )

    # Click announcement element (optional, if required for specific behavior)
    # sampath_announcement.click()  # Uncomment if necessary

    # Wait for download button to appear
    try:
        download_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, "a.dropdown-button.dropdown-toggle[target='_blank']")
            )
        )
    except TimeoutException:
        print("Download button not found within 10 seconds.")
        driver.quit()
        return None

    # Extract and return the download href
    download_href = download_button.get_attribute("href")
    driver.quit()
    return download_href

# Example usage
# url = "https://your_target_website.com"  # Replace with the actual webpage URL
download_href = get_download_href(url)

if download_href:
    print("Download link:", download_href)
else:
    print("Download link not found.")
```
