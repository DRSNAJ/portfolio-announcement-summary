```python
from docling.document_converter import DocumentConverter
from pprint import pprint
from langchain_ollama import OllamaLLM
import os
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
model = OllamaLLM(model="llama3.2:3b")

response = ''

# Specify the directory path
directory_path = "test-files/"

# List all files in the directory
file_names = os.listdir(directory_path)

converter = DocumentConverter()

# Print each file name
for file_name in file_names:
  source = directory_path + file_name # document per local path or URL
  result = converter.convert(source)

  response += "===\n" + model.invoke("Give me concise yet a nuanced summary of this: \n" + result.document.export_to_markdown())

  pprint(response)
  # output: "## Docling Technical Report[...]"
pprint(response)
```

```python
url = "https://www.cse.lk/pages/general-announcements/general-announcements.component.html"
# Create a browser instance
driver = webdriver.Chrome()
# Open a web page
driver.get(url)

select_button = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "button.drop-toggle.flat"))
)
select_button.click()

# search_box = WebDriverWait(driver, 10).until(
#     EC.presence_of_element_located((By.XPATH, "//*[@id='announcement-main-block']/div/div/div[1]/div/div[2]/div[1]/div/div/div/label[1]/input"))
# )
# search_box.click()
# search_box.send_keys("SAMPATH BANK")
# search_box.send_keys(Keys.ENTER)


select_all_checkbox = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "//*[@id='announcement-main-block']/div/div/div[1]/div/div[2]/div[1]/div/div/div/label[2]/input"))
)
select_all_checkbox.click()
driver.execute_script("arguments[0].dispatchEvent(new Event('mouseleave', { bubbles: true }));", select_all_checkbox)
# select_all_checkbox.send_keys(Keys.ENTER)

button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "#announcement-main-block > div > div > div.container-fluid.mb-20.search-bg > div > div:nth-child(2) > form > div.col-lg-12.col-md-12.col-sm-12.col-xs-12 > input"))
)
button.click()

company_name = "CEYLON TOBACCO COMPANY PLC"

# Wait for accordion to be present and visible
accordion = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "accordion"))
)
rows = driver.find_elements(By.CSS_SELECTOR, "#accordion > #accordion > div:nth-child")

print(rows)
# # Wait for rows to be loaded and stable
# rows = WebDriverWait(driver, 10).until(rows)


# Iterate through rows with explicit waits
for row in rows:

    # Wait for company name element to be visible within this row
    company_name_element = WebDriverWait(row, 3).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, "a > div > div.company-name"))
    )
    
    # Compare company names (case-insensitive)
    if company_name_element.text.strip().upper() == company_name.upper():
        print(f"Found company: {company_name}")
        
        # Wait for element to be clickable before clicking
        clickable_element = WebDriverWait(driver, 3).until(
            EC.element_to_be_clickable(row)
        )
        
        clickable_element.click()
```
